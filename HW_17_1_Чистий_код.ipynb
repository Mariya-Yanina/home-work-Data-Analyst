````python
import math                        #використано явний імпорт, покращує читабельність і запобігає конфліктам імен
                                   #додано порожній рядок між імпортом та функцією
def calc_area(r):                  #тіло функціі перенесено на наступний рядок і додано відступ 4 пробіли, змінено назву функції через _ та всі маленькі літери
    return pi * r ** 2             #виділено пробілами оператор "*" та "**"

                                   #між функціями додано 2 порожніх рядки
def calc_circumference(r):
    return 2 * pi * r

radius = 5                         #виділено пробілами оператор "="

print(                             #аргументи функцій розділені пробілами після ком, покращена читабельність
      "Area:", calc_area(radius),
      "Circumference:", calc_circumference(radius)
     )
````

Ваше завдання полягає в тому, щоб змінити код так, аби уникнути використання iterrows() і замість цього застосувати більш ефективне та читабельне рішення відповідно до правил, озвучених у лекції.
````python
import pandas as pd

df = pd.DataFrame({
    'student': ['Alice', 'Bob', 'Charlie', 'David'],
    'math_score': [89, 94, 77, 85],
    'english_score': [92, 88, 79, 85],
    'history_score': [84, 92, 81, 89]
})
````

````python
averages = []
for _, row in df.iterrows():
    averages.append((row['math_score'] + row['english_score'] + row['history_score']) / 3)

df['average_score'] = averages
````

````python
df['average_score'] = df[
    ['math_score', 'english_score', 'history_score']
].mean(axis=1)
````

````python
def calculate_variance(numbers):
    sum_values = 0
    for number in numbers:
        sum_values += number

    mean = sum_values / len(numbers)

    sum_squares = 0
    for number in numbers:
        sum_squares += number ** 2

    mean_squares = sum_squares / len(numbers)

    return mean_squares - mean ** 2
````
